Currently MTN Cote d'Ivoire would like to upgrade its technology infrastructure for its mobile users in Ivory Coast. Studying the given dataset, how does MTN Cote d'Ivoire go about the upgrade of its infrastructure strategy within the given cities?

```
# importing the library
import pandas as pd
```

**LOADING THE DATA**

```
# Loading TELCOM_DATASETS files

# Column headers
column_headers = ['PRODUCT', 'VALUE', 'DATETIME',
                  'CELL_ON_SITE', 'DW_A_NUMBER_INT', 'DW_B_NUMBER_INT',
                  'COUNTRY_A', 'COUNTRY_B', 'CELL_ID', 'SITE_ID']

telcom_dataset1 = pd.read_csv("Telcom_dataset.csv", skiprows=1, header=None, names=column_headers)
telcom_dataset2 = pd.read_csv("Telcom_dataset2.csv", skiprows=1, header=None, names=column_headers)
telcom_dataset3 = pd.read_csv("Telcom_dataset3.csv", skiprows=1, header=None, names=column_headers)
```

**DATA PREPARATION/CLEANING**

```
# Combining the data into a single Dataframe
df = pd.concat([telcom_dataset1, telcom_dataset2, telcom_dataset3])
```

```
# Dropping any duplicate data
df = df.drop_duplicates()
```

```
# Checking for data type
df.info()
```

```
# Resetting the index
df.reset_index(drop=True, inplace=True)
```

```
# Checking our data
df
```

```
# Loading CELL_GEO
geo_data = pd.read_csv("cells_geo.csv", delimiter=';')
```

```
# Viewing the data
geo_data
```

```
# Removing any duplicate values
geo_data = geo_data.drop_duplicates()
```

```
# General information about the DataFrame
geo_data.info()
```

```
# Dropping the 'Unnamed: 0' column
geo_data = geo_data.drop(['Unnamed: 0'], axis = 1)

geo_data
```

```
# Checking the columns in CDR Dataframe
df.columns
```

```
# Checking the colimns in GEO DATA Dataframe
geo_data.columns
```

```
# Renaming a column in GEO DATA
geo_data.rename(columns = {'SITE_CODE':'SITE_ID'}, inplace = True) 

# Checking the name change
geo_data.columns
```

```
# Combining the two data into one mega file
combined_df = pd.merge(df, geo_data, on='SITE_ID', how='inner')

combined_df
```

```
# Converting the DATETIME column to type datetime
combined_df['DATETIME'] = pd.to_datetime(combined_df['DATETIME'])
```

```
# Resetting DATETIME to index
combined_df.set_index(["DATETIME"], inplace = True, append = True, drop = True) 
```

```
# Exploring the data types in the combined file
combined_df.info()
```
